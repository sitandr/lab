```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("../data/low-pass.csv", sep=";", comment="#")
```

```python
freq_n, voltage_n = df
df = df.sort_values(freq_n)
freq, voltage = df[freq_n], df[voltage_n]
freq *= 1e3
voltage/=3
```

```python
C = 0.22e-6
R = 1e3

ω = np.arange(min(freq), max(freq), 1) * 2*np.pi
theory = 1/(R**2*(ω**2*C**2) + 1)**0.5
```

```python

%matplotlib widget
plt.loglog()
plt.grid(ls='--', c='0.8')
plt.scatter(freq, voltage, s=6)
plt.plot(ω/2/np.pi, theory/2**0.5)
plt.xlabel("Частота сигнала, Гц")
plt.ylabel("Доля пропускаемого сигнала")
pass
```
