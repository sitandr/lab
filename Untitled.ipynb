```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit, minimize
from IPython.display import display
from scipy.interpolate import UnivariateSpline
```

```python
FR_0 = pd.read_csv('0Ω', sep=';', comment='#')
FR_30 = pd.read_csv('30Ω', sep=';', comment='#').sort_values('ν, kHz').reset_index(drop=True)
# zatuh = pd.read_csv('zatuh', sep=';', comment='#')
m_0 = np.argmax(FR_0['U, mV'])
U_0, f_0 = FR_0['U, mV'][m_0], FR_0['ν, kHz'][m_0]
display(U_0, f_0)

print()

m_30 = np.argmax(FR_30['U, mV'])
U_30, f_30 = FR_30['U, mV'][m_30], FR_30['ν, kHz'][m_30]
display(U_30, f_30)
# FR_30['U, mV']
```

```python
%matplotlib widget
plt.plot(FR_30['ν, kHz'], FR_30['U, mV'], '-o', markersize=4)
plt.xlabel('$f$, Гц')
plt.ylabel('$U$, мВ')
plt.savefig('orig_FR_30.png')
```
