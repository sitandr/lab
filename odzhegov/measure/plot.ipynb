```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import pandas as pd
import re
```

```python
from io import StringIO
from os import listdir
from os.path import isfile

thefiles = [f for f in listdir('.') if isfile(f) and f.endswith(".csv")]
```

```python
for f in thefiles:
    text = open(f).read()
    text = re.sub("\[.*\]", "", text)
    df = pd.read_csv(StringIO(text), delimiter=";", comment="#", header=None)
    fig, ax = plt.subplots(1, 1)
    #ax.set_title(f[:-4])
    ax.minorticks_on()
    ax.plot(df[0], df[1])
    ax.set_xlabel("λ, нм")
    fig.savefig(f[:-4] + '.jpg', dpi=100)
    fig.show()
```

```python
f = "no-glass halogen lamp.csv"
text = open(f).read()
text = re.sub("\[.*\]", "", text)
default = pd.read_csv(StringIO(text), delimiter=";", comment="#", header=None)

for f in ['after_green_filter.csv', 'green glass.csv', 'Schott-glass filter halogen lamp.csv']:
    text = open(f).read()
    text = re.sub("\[.*\]", "", text)
    this = pd.read_csv(StringIO(text), delimiter=";", comment="#", header=None)
    df = this
    df[1] = df[1].apply(lambda x: 0 if x < 0.1 else x)
    df[1][1045] = 0
    df[1] /= default[1]
    
    fig, ax = plt.subplots(1, 1)
    #ax.set_title(f[:-4])
    ax.minorticks_on()
    ax.plot(df[0], df[1])
    ax.set_xlabel("λ, нм")
    fig.savefig(f[:-4] + '_normed.jpg', dpi=100)
    fig.show()
```
