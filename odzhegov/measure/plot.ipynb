```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import pandas as pd
import re
```

```python
from io import StringIO
from os import listdir
from os.path import isfile

thefiles = [f for f in listdir('.') if isfile(f) and f.endswith(".csv")]
```

```python
for f in thefiles:
    text = open(f).read()
    text = re.sub("\[.*\]", "", text)
    df = pd.read_csv(StringIO(text), delimiter=";", comment="#", header=None)
    fig, ax = plt.subplots(1, 1)
    #ax.set_title(f[:-4])
    ax.minorticks_on()
    ax.plot(df[0], df[1])
    ax.set_xlabel(", ")
    fig.savefig(f[:-4] + '.jpg', dpi=100)
    fig.show()
```
